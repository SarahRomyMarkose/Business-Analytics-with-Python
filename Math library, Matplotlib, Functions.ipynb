```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,3,5,7,9,2,4,6,8]

plt.plot(x,y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.scatter(x,y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.scatter(x,y, marker="*")  # Here, '*'/asterix refers to star shape.
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.plot(x,y, marker="D") # Here, 'D' refers to Diamond shape.
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.plot(x,y, marker="D", ms=20)  # 'ms' determines the size of induvidual plots.
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.plot(x,y, marker="D", ms=20, mec="r")   # 'mec' determines the outer color of induvidual plots.
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.plot(x,y, marker="D", ms=20, mec="r", mfc="g")  # 'mfc' determines the color of induvidual plots.
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```
