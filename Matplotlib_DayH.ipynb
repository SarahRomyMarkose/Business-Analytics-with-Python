```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('TUP.csv')
```

```python
x=dataset.iloc[:,:-1].values
```

```python
y=dataset.iloc[:,-1].values
```

```python
from sklearn.svm import SVR
regressor=SVR()
regressor.fit(x,y)
```

```python
y_pred=regressor.predict(x)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,regressor.predict(x),color='blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor=SVR(kernel='rbf')
regressor.fit(x,y)
y_pred=regressor.predict(x)
plt.scatter(x,y,color='red')
plt.plot(x,regressor.predict(x),color='blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```
