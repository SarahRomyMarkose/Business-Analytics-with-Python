```python
x=int(input("Income:")) #Case2
if x<=250000:
    print("No Tax")
elif x<=500000:
    T=(x-250000)*0.05
    print(T)
elif x<=750000:
    T=((x-500000)*0.1)+12500
    print(T)
elif x<=1000000:
    T=((x-750000)*0.15)+37500
    print(T)
elif x<=1250000:
    T=((x-1000000)*0.20)+75000
    print(T)
elif x<=1500000:
    T=((x-1250000)*0.25)+125000
    print(T)
elif x>1500000:
    T=((x-1500000)*0.30)+187500
    print(T)
```
