```python
df=pd.read_csv("Supermarket_sales.csv")
new_data=[]
for x in df.index:
    if(df.loc[x,"Productline"]=="Health and beauty"):
        if (df.loc[x,"Gender"]=="Male"):
            new_data.append(df.loc[x,"InvoiceID"])
len(new_data)
```
